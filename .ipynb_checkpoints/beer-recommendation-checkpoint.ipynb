### Preprocess reviews

```python
def extract_main(review):
    """
    extract the main part of review from each review
    args:
        review to be parsed (str)
    return:
        main body of the review (str)
    """
    return review.split("\n\n")[1]

df_master['main_review'] = df_master['product_review'].map(extract_main)
```

```python
df_master[:5]
```

```python
# Prep each main review for attribute counts
import nltk
# tokenization
nltk.download('punkt')
from nltk.tokenize import word_tokenize
# stopword
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
# the class example did not lemmatize
```

```python
def preprocessreview(review):
    """
    tokenize and remove stopwords from review for counting
    args:
        main review (string)
    returns:
        a list of words
    """
    word_tokens = word_tokenize(review) # tokenize words
    filtered_sentence = [word.lower() for word in word_tokens if word not in stop_words] # remove stopwords
    return filtered_sentence

df_master['processed_review'] = df_master['main_review'].map(preprocessreview)
```

```python
df_master[:2]
```

### Choose three attributes (by count) to recommend beers on

```python
# conduct word frequency analysis of the interested words
interested_words = {'aggressive':0, 'balanced':0,'complex':0,'crisp':0,'fruity':0,
                   'hoppy':0,'malty':0,'robust':0}
for review in df_master['processed_review']:
    for word in review:
        if word in interested_words:
            interested_words[word] += 1
```

```python
sorted(interested_words.items(), key=lambda x: x[1], reverse = True)
```

### Similarity Analysis
Finding which reviews are "most similar" to the attributes chosen (i.e., which reviews mention the chosen attributes, whether directly or with synonyms)

```python
#!pip install spacy
#!python -m spacy download en
#!python -m spacy download en_core_web_md
```

```python
import spacy
nlp = spacy.load("en_core_web_md")  #en_core_web_sm
```

```python
# test on calculating similarity score of two strings
string1 = nlp(df_master['main_review'][0])
string2 = nlp('balanced complex fruity')
# for word1 in string1:
#     for word2 in string2:
#         print(word1, word2, word1.similarity(word2))
string1.similarity(string2)
```

```python
import numpy as np
def computesimilarity(attribute_file, review):
    """
    computes the spacy-based similarity score between the two strings given
    args:
        attributes (txt file): desired attributes the user wishes to calculate each review with in a txt file
        review (str): the review in question
    return:
        similarity score (float) between the attributes and the review
    """
    attributes = ''
    for attribute in pd.read_csv(attribute_file,header = None)[0].values:
        attributes += attribute + ' '
    attributes = nlp(attributes.rstrip())
    review = nlp(review)
    return attributes.similarity(review)
```

```python
computesimilarity("beer_attributes.txt", df_master['main_review'][0])
```

```python
import functools
computesimilarity_partial = functools.partial(computesimilarity, "beer_attributes.txt")
df_master['similarity_score'] = df_master['main_review'].map(computesimilarity_partial)
```

#### Top 300 reviews by similarity scores

```python
df_master.sort_values(by = ['similarity_score'], ascending = False)[:300]
```

```python
# Store the 300 reviews with the highest similarity scores in an output file
df_similarity = df_master[['product_name','main_review','similarity_score']].sort_values(by='similarity_score', ascending = False)[:300]
df_similarity.rename(columns = {'main_review':'product_review'},inplace = True)
df_similarity.to_csv("beer_similarity.csv", index=False)
```

### Perform Sentiment Analysis on the 300 Reviews

```python
from nltk.sentiment import vader
```

```python
beer_sent = vader.SentimentIntensityAnalyzer()
def get_sentiment(review):
    """
    get compound sentiment score of a beer review
    args:
        review in question (str)
    return:
        sentiment score (float)
    """
    return beer_sent.polarity_scores(review)['compound']
```

```python
df_similarity['sentiment_score'] = df_similarity['product_review'].map(get_sentiment)
```

#### Sort reviews by sentiment scores

```python
df_similarity.sort_values(by='sentiment_score', ascending = False)
```

### Recommend 3 Beers based on highest average sentiments

```python
df_similarity.groupby('product_name').mean().sort_values(by='sentiment_score', ascending = False).iloc[:3]
```

We recommend **Mocha Wednesday, Grey Monday, and Cutting Tiles (Galaxy)** for a beer drinker who is looking for a balanced, complex, and fruity beer. (All the beers in our sample size of 300 already have decently high similarity score with the three attributes.)

Without the above analysis, you may not think of Mocha Wednesday as being a fruity beer. However, we sampled and analyzed the reviews in detail. And the beer is actually being reviewed with dark fruits such as cherries and raisins. Grey Monday is similar to Mocha Wednesday (both are stouts) and exhibit similar characteristics. Cutting Tiles would be your more traditional "fruity" beer. It is a double IPA with hints of nectarines, peaches, and honey.
